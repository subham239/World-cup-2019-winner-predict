```python
results=pd.read_csv('D:/study material/world-cup-predictor/datsets/results.csv')
```

```python
results.head()
```

```python
#data cleaning


#All matches played by India
df=results[(results['Team_1']=='India') | (results['Team_2']=='India')] 

```

```python
df
```

```python
india=df.iloc[:]
```

```python
#store India's results of any particular year for future purpose  

year=[]
for row in india['date']:
    if len(row)== 8:
        year.append(int(row[6:]))
    else:
        year.append(int(row[7:]))

india['match_year']=year
india_2010=india[india.match_year==10]
india_2010.count()
```

```python
#analysis of teams those are participating in worldcup

worldcup_teams=['England','South Africa','India','Pakistan','Bangladesh','Afghanistan','New Zealand','Australia','West Indies','Sri Lanka']
df_teams1=results[results['Team_1'].isin(worldcup_teams)]
df_teams2=results[results['Team_2'].isin(worldcup_teams)]
df_teams=pd.concat((df_teams1,df_teams2))
df_teams.drop_duplicates()
df_teams
```

```python
df_teams_2010=df_teams.drop(['date','Margin','Ground'],axis=1)
df_teams_2010
```

```python
#assigning the winning team (if first team wins then assign winning team as 1 and if second team wins assign winning team as 2)


df_teams_2010=df_teams_2010.reset_index(drop=True)

df_teams_2010.loc[df_teams_2010.Winner == df_teams_2010.Team_1,'winning_team']=1
df_teams_2010.loc[df_teams_2010.Winner == df_teams_2010.Team_2,'winning_team']=2

df_teams_2010 = df_teams_2010.drop(['winning_team'], axis=1)

df_teams_2010.head()
```

```python
final=pd.get_dummies(df_teams_2010,prefix=['Team_1','Team_2'],columns=['Team_1','Team_2'])
```

```python
X=final.drop(['Winner'],axis=1)
y=final["Winner"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
final.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
```

```python
rf = RandomForestClassifier(n_estimators=100, max_depth=20,random_state=0) 

#print("Training set accuracy: ", '%.3f'%(score)) print("Test set accuracy: ", '%.3f'%(score2))
```

```python
rf.fit(X_train, y_train)
```

```python
score = rf.score(X_train, y_train)
score2 = rf.score(X_test, y_test)
```

```python
print("training set accuracy")
score
```

```python
print("testing set accuracy")
score2
```
